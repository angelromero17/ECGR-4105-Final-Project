```python
data = dataset.copy()

value_count = data['Specialization'].value_counts()

def map_to_other_specialization(var):
    
    if var in value_count[value_count<=10]:
        return 'other'
    else:
        return var
    
    dataset['Specialization'] = dataset.Specialization.apply(map_to_other_specialization)
```

```python
plt.figure(figsize = (16, 8))
total = float(len(dataset))
ax = sns.countplot(x='Specialization',data=dataset)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 1,
            '{:1.2f}%'.format((height/total) * 100),
            ha="center",fontsize=10) 
plt.xticks(rotation = 90)
plt.show()
```

```python
data = dataset.copy()

value_count = data['GraduationYear'].value_counts()

def map_to_other_GraduationYear(var):
    
    if var in value_count[value_count<=10]:
        return 'other'
    else:
        return var
    
    dataset['GraduationYear'] = dataset.GraduationYear.apply(map_to_other_GraduationYear)
```

```python
plt.figure(figsize = (16, 8))
total = float(len(dataset))
ax = sns.countplot(x='GraduationYear',data=dataset)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 1,
            '{:1.2f}%'.format((height/total) * 100),
            ha="center",fontsize=10) 
plt.xticks(rotation = 90)
plt.show()
```

```python
X = dataset.drop('Salary', axis=1)
y = dataset['Salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=50)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
y_pred_linear_reg = linear_reg.predict(X_test)

linear_reg_r2_score = linear_reg.score(X_test, y_test)
```

```python
errors_reg = y_test - y_pred_linear_reg

plt.figure(figsize = (10,6))
sns.displot(errors_reg)
plt.show()
```
